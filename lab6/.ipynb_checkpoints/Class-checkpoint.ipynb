```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt       
```

```python
data_path = 'spambase.data'
columns = [f'feature_{i}' for i in range(1, 58)] + ['label']
data = pd.read_csv(data_path, header=None, names=columns)
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
class StandardScalerCustom:
    def fit_transform(self, X):
        self.mean = X.mean(axis=0)
        self.std = X.std(axis=0)
        return (X - self.mean) / self.std
scaler = StandardScalerCustom()
X = scaler.fit_transform(X)
```

```python
def train_test_split_custom(X, y, test_size=0.3, random_state=None):
    np.random.seed(random_state)
    indices = np.arange(len(y))
    np.random.shuffle(indices)
    split_idx = int(len(y) * (1 - test_size))
    train_indices = indices[:split_idx]
    test_indices = indices[split_idx:]
    return X[train_indices], X[test_indices], y[train_indices], y[test_indices]
X_train, X_test, y_train, y_test = train_test_split_custom(X, y, test_size=0.3, random_state=42)
```

```python
class LogisticRegressionCustom:
    def __init__(self, learning_rate=0.01, max_iter=1000):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0
        for _ in range(self.max_iter):
            linear_model = np.dot(X, self.weights) + self.bias
            predictions = self.sigmoid(linear_model)
            dw = (1 / len(y)) * np.dot(X.T, (predictions - y))
            db = (1 / len(y)) * np.sum(predictions - y)
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    def predict_proba(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self.sigmoid(linear_model)
    def predict(self, X, threshold=0.5):
        probabilities = self.predict_proba(X)
        return (probabilities >= threshold).astype(int)
```

```python
def calculate_metrics(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))

    accuracy = (tp + tn) / len(y_true)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return accuracy, precision, recall, specificity, f1_score
```

```python
model = LogisticRegressionCustom(learning_rate=0.1, max_iter=1000)
model.fit(X_train, y_train)
def evaluate_thresholds(model, X_test, y_test):
    thresholds = np.arange(0.1, 1.0, 0.1)
    metrics = []
    for threshold in thresholds:
        y_pred = model.predict(X_test, threshold)
        accuracy, precision, recall, specificity, f1_score = calculate_metrics(y_test, y_pred)
        metrics.append((threshold, accuracy, precision, recall, specificity, f1_score))
    return metrics
metrics = evaluate_thresholds(model, X_test, y_test)
```

```python
for threshold, accuracy, precision, recall, specificity, f1_score in metrics:
    print(f"Threshold: {threshold:.1f}, Accuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, Specificity: {specificity:.2f}, F1 Score: {f1_score:.2f}")
```

```python
def plot_roc_curve(model, X_test, y_test):
    y_prob = model.predict_proba(X_test)
    thresholds = np.linspace(0, 1, 100)
    tpr_list = []
    fpr_list = []
    for threshold in thresholds:
        y_pred = (y_prob >= threshold).astype(int)
        tp = np.sum((y_test == 1) & (y_pred == 1))
        tn = np.sum((y_test == 0) & (y_pred == 0))
        fp = np.sum((y_test == 0) & (y_pred == 1))
        fn = np.sum((y_test == 1) & (y_pred == 0))
        tpr = tp / (tp + fn) if (tp + fn) > 0 else 0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr_list, tpr_list, label="ROC Curve", color="blue")
    plt.plot([0, 1], [0, 1], "k--", label="Random Guess")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.show()
plot_roc_curve(model, X_test, y_test)
```
